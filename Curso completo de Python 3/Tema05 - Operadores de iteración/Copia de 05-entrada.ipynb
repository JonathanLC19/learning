# 99 Bottles of Rum on the Wall

De tantas solicitudes que ha recibido, Pyratilla no cabe en sí de la emoción y se pone a cantar

<ul type = "none">
<li>"99 bottles of rum on the wall, 99 bottles of rum. </li>
<li>Take one down, pass it around, 98 bottles of rum on the wall. </li>
<li>98 bottles of rum on the wall, 98 bottles of rum. </li>
<li>Take one down, pass it around, 97 bottles of rum on the wall..." </li>
</ul>


Ayuda a Pyratilla a cantar la canción al completo con un bucle `while`.

```
n = 99

while n >= 1:
    print("{} bottles of rum on the wall, {} bottles of rum.\nTake one down, pass it around, {} bottles of rum on the wall.\n".format(n , n , n-1))
    n -= 1
```

Ayuda a Pyratilla a cantar la canción al completo con un bucle `for`.

```
n = 99

for x in range(n):
    print("{} bottles of rum on the wall, {} bottles of rum.\nTake one down, pass it around, {} bottles of rum on the wall.\n".format(n , n , n-1))
    n -= 1
```
